# Synth dataset

```python
def gen_data(churn_rate=0.1, base_customer_size=1000, time=50):
    dat = geom.rvs(churn_rate, size=base_customer_size)
    return dat

def survival(dat):
    survival_curve = []
    for i in range(min(dat),max(dat)+1):
        #number of customer stil active at each month i
        survival_curve += np.sum(dat>=i),
    return np.array(survival_curve)
```

# Churn Rate change over time 

```python
n_cohort = 5
base_churn = 0.2
churn_increment = 0.33

actual_rates = []

dat = []
for i in range(n_cohort):
    churn_rate = base_churn + base_churn * churn_increment * i
    actual_rates += churn_rate,
    cohort = geom.rvs(churn_rate, size=1000)
    dat.append(cohort)
    
dat = np.array(dat)

dat.shape
```

```python
actual_rates
```

We'll recover this rates in the next sections

## Separate model for each rate
Let's model each rate separately for each cohort

```python

with pm.Model() as fix_churn:
    rate = pm.Beta('rate',1,1,shape=len(dat))
    for i in range(len(dat)):
        _ = pm.Geometric('lkl'+str(i),p=rate[i],observed=dat[i])
    
    trace = pm.sample(2000,tune=3000)

pm.plot_trace(trace);    
pm.plot_posterior(trace,var_names=['rate'],ref_val=actual_rates);
```

## Rate changes over times 

Now that we can see there is some pattern in how our rate changes. We theorize that there is some sort of a trend in our rate over time. This is of course latent. We'll model this change over time as a linear function.

Later on, if we see our theory of a linear latent model for rates does not make sense, as much as other model of changes. We can fit those competing models and see which one fit our data beter. 

```python

with pm.Model() as fix_churn:
    base_rate = pm.Beta('rate',1,1)
    coeff = pm.Normal('coeff',0,0.5)
    
    for i in range(len(dat)):
        rate = base_rate * (1 + coeff * i)
        _ = pm.Geometric('lkl'+str(i), p=rate, observed=dat[i])
    
    trace = pm.sample(2000,tune=3000,)

pm.plot_trace(trace);    
pm.plot_posterior(trace,var_names=['rate','coeff'],ref_val=[base_churn, churn_increment]);
```

```python
pm.summary(trace)
```

```python
pm.autocorrplot(trace[::]);
```

## Censor the data to simulate the real world

```python
MAX_DATA_LEN  = 10

dat_cen = []

for i in range(len(dat)):
    dat_i = dat[i].copy()
    mask = dat_i > (MAX_DATA_LEN - i)
    dat_i[mask] = MAX_DATA_LEN - i
    dat_cen.append(dat_i)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.histplot(dat[0],label='Original histogram')

plt.legend()
plt.subplot(1,2,2)
sns.histplot(dat_cen[0],label='Censored histogram')
plt.legend();
```

## Method1: throw away censored data
- only choose to model uncensored data points ( no spikes at the right)
- we loose some information, but hopefully the remainning data is sufficient
- this method is sometime called truncation

```python

with pm.Model() as fix_churn:
    base_rate = pm.Beta('rate',1,1)
    coeff = pm.Normal('coeff',0,0.1)
    
    for i in range(len(dat)):
        
        rate = base_rate * (1 + coeff * i)
        
        dat_trunc = dat[i].copy()
        mask = dat_trunc < dat_trunc.max()
        dat_trunc  = dat_trunc[mask]
        
        lkl = pm.Geometric('lkl'+str(i),p=rate,observed=dat_trunc)
    
    trace = pm.sample(2000,tune=3000)

pm.plot_trace(trace,var_names=['rate','coeff']);    
pm.plot_posterior(trace,var_names=['rate','coeff'],ref_val=[base_churn, churn_increment]);
```

```python
plt.figure(figsize=(15,3))
for i in range(len(dat)):
    rate_i = trace['rate'] * (1 + trace['coeff'] * i)
    sns.kdeplot(rate_i,label='month'+str(i))
    true_rate = base_churn * (1+ churn_increment * i)
    plt.axvline(true_rate,color='black',alpha=0.3)
```

The model seems to deal ok with the some data removed. The right censored part is small in this model. However, there are situations when we want to model some churn rates with a lot of censored datapoints (early in the cohort cycle), then throwing away datapoints is detrimental. In that case, we can use these censored datapoints to guide the model by supplying it with supplemental information from these censored points

## Censored data method  Imputation: truncate data at the threshold and model them separately
- Adding bounded likelihood to guide the MCMC machinery

```python

with pm.Model() as fix_churn:
    base_rate = pm.Beta('rate',1,1)
    coeff = pm.Normal('coeff',0,0.1)
    
    for i in range(len(dat)):
        rate = base_rate * (1 + coeff * i)
        dat_trunc = dat[i].copy()
        mask = dat_trunc < dat_trunc.max()
        dat_trunc  = dat_trunc[mask]
        
        #there is some extra information this way sir!
        BGeometric = pm.Bound(pm.Geometric, lower=dat_trunc.max())
        _ = BGeometric('bgeo'+str(i),p=rate,shape=(sum(~mask),))
        
        lkl = pm.Geometric('lkl'+str(i),p=rate,observed=dat_trunc)
    
    trace = pm.sample(2000,tune=3000)

pm.plot_trace(trace,var_names=['rate','coeff']);    
pm.plot_posterior(trace,var_names=['rate','coeff'],ref_val=[base_churn, churn_increment]);
```

```python
plt.figure(figsize=(15,3))
for i in range(len(dat)):
    rate_i = trace['rate'] * (1 + trace['coeff'] * i)
    sns.kdeplot(rate_i,label='month'+str(i))
    true_rate = base_churn * (1+ churn_increment * i)
    plt.axvline(true_rate,color='black',alpha=0.3)
```

## Method 3: separate density for censored data
- This is equivalent to the one above, but do not need to models many extra variables

```python
import theano.tensor as tt
```

```python

def tt_logsf(x, p):
    """ Geometric log complementary cdf (or log survival) in theano
        https://github.com/scipy/scipy/blob/main/scipy/stats/_discrete_distns.py#L433
    """
    k = tt.floor(x)
    return k * tt.log1p(-p)
```

```python

with pm.Model() as fix_churn:
    base_rate = pm.Beta('rate',1,1)
    coeff = pm.Normal('coeff',0,0.1)
    
    for i in range(len(dat)):
        rate = base_rate * (1 + coeff * i)
        
        dat_trunc = dat[i].copy()
        mask = dat_trunc < dat_trunc.max()
        
        dat_trunc  = dat_trunc[mask]
        lkl = pm.Geometric('lkl'+str(i),p=rate,observed=dat_trunc)
        right_censored = pm.Potential('right_cen'+str(i),
                                        np.sum(~mask) * tt_logsf(x=dat_trunc.max(),p=rate))
                                     
    trace = pm.sample(2000,tune=2000)

pm.plot_trace(trace,var_names=['rate','coeff']);    
pm.plot_posterior(trace,var_names=['rate','coeff'],ref_val=[base_churn, churn_increment]);
```

```python
plt.figure(figsize=(15,3))
for i in range(len(dat)):
    rate_i = trace['rate'] * (1 + trace['coeff'] * i)
    sns.kdeplot(rate_i,label='month'+str(i))
    true_rate = base_churn * (1+ churn_increment * i)
    plt.axvline(true_rate,color='black',alpha=0.3)
```
